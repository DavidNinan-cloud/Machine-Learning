```python
# TRAIN, TEST SPLIT
from sklearn.model_selection import train_test_split
#Getting all Coulmn names
USA_Housing.columns

```

```python
X=USA_Housing[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]

y=USA_Housing['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30)
```

```python
#Linear Regression Algorithm
from sklearn.linear_model import LinearRegression

lm_model = LinearRegression()
lm_model.fit(X_train, y_train)
lm_pred=lm_model.predict(X_test)
```

```python

from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, lm_pred))
print('MSE:', metrics.mean_squared_error(y_test, lm_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, lm_pred)))
```
